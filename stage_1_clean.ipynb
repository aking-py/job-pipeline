# Job Market Data Pipeline

**An automated Python data pipeline that performs cleansing and analysis on LinkedIn job postings data from 2023-2024. Built primarily using pandas and numpy, the pipeline uncovers hring trends, in-demand skills and salary patterns across the data industry.**

**Imports**

```python
import numpy as np
import pandas as pd
```

Imported the two libraries pandas and numpy; pandas, being the tool used for loading, cleaning and manipulating data while numpy is used for numerical and statistical analysis. 

**Loading Data**

```python
df = pd.read_csv("raw/postings.csv")

df.head(10)
```

```python
print(df.shape)
```

```python
print(df.columns)
```

**Locating missing values**

```python
print(df.isnull().sum())
```

**My next step was to find missing values within my dataframe using the isnull() 
function. This displays each column name and how many missing values are in those 
columns. This gives me a greater understanding of which data points aren't worth 
analysing.**

**Dropping columns**

```python
df = df [['title', 'company_name', 'location', 'formatted_work_type', 'remote_allowed', 'formatted_experience_level', 'min_salary', 'max_salary', 'normalized_salary', 'views', 'applies', 'listed_time']]

print(df.shape)
```

**Here, I have selected only the relevant columns for analysis and dropped the rest 
from the dataframe. I identified which columns to remove by using the isnull().sum() 
function to check for missing values. Columns with excessive missing values such as 
closed_time and skills_desc were removed, as were irrelevant columns such 
as job_id, posting_domain and application_url.**

**Checking data types**

```python
df.dtypes
```

**Here, I have checked the data types of each column to ensure they are accurate 
and best suited for analysis. Title, company_name and location are all object 
(string/text) columns which is correct. However, columns such as min_salary, 
views and applies are float64 (decimal) which is not ideal — these will be 
converted to integers as you cannot have half a salary or half a view. 
Additionally, remote_allowed will be converted to a boolean (True/False) and 
listed_time will be converted to a datetime format.**

**Converting data types**

```python
df = df.fillna(0).astype({'min_salary':'int','max_salary':'int','normalized_salary':'int','views':'int','applies':'int'})
```

**Here, my aim was to convert several column data types from float64 to integers. 
These columns were min_salary, max_salary, normalized_salary, views and applies. 
Before converting, I first had to handle the NULL values within these columns by 
filling them with the value '0' using the fillna() method, chained with astype() to perform 
the conversion in a single line. Skipping this step would result in an error, as 
Python cannot convert NULL values to another data type.**

**Converting column to datetime**

```python
df['listed_time'] = pd.to_datetime(df['listed_time'], unit = 'ms')
```

```python
df['listed_time']
```

**Here, I converted the listed_time column from a Unix timestamp to a readable 
datetime format. During this step I encountered a debugging issue — the raw values 
were stored as scientific numbers (e.g. 1.71e+12), representing milliseconds since 
1st January 1970. Initially the dates were converting incorrectly, displaying 1970 
dates. By inspecting the raw values I identified the issue and resolved it by using 
pd.to_datetime() with the unit='ms' argument, which correctly converted the values 
to 2023-2024 dates.**

**Checking for and dropping duplicate rows**

```python
print(sum(df.duplicated()))
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

**Final check before analysis**

```python
df
```

**Changing the integers ("0") to "Unknown" in our string/text columns**

```python
df['company_name'] = df['company_name'].replace(0,'Unknown')
df['location'] = df['location'].replace(0,'Unknown')
df['formatted_experience_level'] = df['formatted_experience_level'].replace(0,'Unknown')
```

**Converting remote_allowed column to boolean**

```python
df['remote_allowed'] = df['remote_allowed'].astype('bool')
```

```python
df["remote_allowed"].dtype
```

```python
df.head(10)
```

**As you can see, I've converted the remote_allowed column to boolean instead of float64. Previously showing either 0.0 or 1.0, the column values now show either True or False.**

```python
print(df.shape)
print(df.dtypes)
df.head()
```

**Within Stage 1, I completed the initial cleaning process of the raw LinkedIn 
job postings dataset. This involved loading the dataset and inspecting the column 
names, data types and missing values. Irrelevant and incomplete columns were dropped 
from 31 down to 12. Null values were handled using fillna() and incorrect 0 values 
in text columns were replaced with "Unknown". Data types were converted to their 
appropriate formats including integers, boolean and datetime. Finally, 803 duplicate 
rows were removed, leaving a cleaned dataset of 123,046 rows.**
