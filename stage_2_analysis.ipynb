### Salary Landscape ###

```python
print("Average salary =", np.mean(df['normalized_salary']))
print("Median salary =", np.median(df['normalized_salary']))
print("Highest salary =", np.max(df['normalized_salary']))
print("Lowest salary =", np.min(df['normalized_salary']))
```

```python
df_salaries = df[(df['normalized_salary'] > 20000) & (df['normalized_salary'] < 200000)] 
```

```python
print(df_salaries.shape)
```

```python
print((df['normalized_salary'] == 0).sum())
```

```python
print("Average salary =", np.mean(df_salaries["normalized_salary"]))
print("Median salary =", np.median(df_salaries["normalized_salary"]))
print("Highest salary =", np.max(df_salaries["normalized_salary"]))
print("Lowest salary =", np.min(df_salaries["normalized_salary"]))
```

**Initially I used the normalized_salary column to calculate average, median, highest and lowest salaries. However the results were inaccurate — the median and lowest salaries were 0, indicating an excess of null values that had been filled with 0, and the highest salary was \$535,600,000 which is clearly an anomaly or data entry error. To address this I filtered the dataset to only include salaries between \\$20,000 and \\$200,000, removing outliers and keeping the data realistic. This left 33,752 valid salary rows. The filtered results show salaries ranging from \\$20,782 to \\$199,950, with an average of \\$87,562 and a median of \\$79,040. The \\$8,500 gap between the two suggests that higher paying roles are pulling the average upward.**

### Remote vs On-Site roles
**Here, I analyse the split between remote and on-site job postings across the dataset.**

```python
print(df.columns)
```

```python
print(df["remote_allowed"].value_counts())
```

**Using the value_counts() method on the remote_allowed column, I uncovered that 
107,830 roles (87.6%) are on-site, whilst 15,216 roles (12.4%) are remote. 
This is a notable finding — despite the shift towards remote working following 
COVID-19, the overwhelming majority of roles advertised during 2023-2024 still 
require physical presence.**

### Top Hiring Companies ###
**Here, I analyse the top hiring companies amongst the job postings in this period.**

```python
filtered_df = df[df["company_name"] != "Unknown"]
print(filtered_df["company_name"].value_counts().head(10))
```

**Excluding "Unknown" values, Liberty Healthcare and Rehabilitation Services leads 
the top 10 with 1,085 job postings, reflecting the high demand in the healthcare 
sector. The Job Network and J. Galt, both recruitment agencies posting on behalf 
of multiple companies, follow in 2nd and 3rd place. This theme continues with 
TEKsystems, Insight Global and Dice also appearing in the top 10, all of which 
are staffing agencies. Capital One and Lowe's stand out as the only recognisable 
direct employers in the list, suggesting that a significant portion of hiring is 
conducted through recruitment agencies rather than directly.**
